```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('housing_area_price.csv')

# Display the first few rows of the dataset
print(df.head())

# Visualize the data
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (in $)')
plt.scatter(df['area'], df['price'], color='red', marker='+', label='Data Points')
plt.legend()
plt.show()

# Split the data into features (X) and target (y)
X = df[['area']]
y = df['price']

# Create a Linear Regression model
reg = LinearRegression()
reg.fit(X, y)

# Predict the price of a home with area = 3300 sq ft
area_3300 = np.array([[3300]])
predicted_price_3300 = reg.predict(area_3300)
print(f"Predicted price of a 3300 sq ft home: ${predicted_price_3300[0]:,.2f}")

# Predict the price of a home with area = 5000 sq ft
area_5000 = np.array([[5000]])
predicted_price_5000 = reg.predict(area_5000)
print(f"Predicted price of a 5000 sq ft home: ${predicted_price_5000[0]:,.2f}")

# Display the slope (coefficient) and intercept
print(f"Coefficient (slope): {reg.coef_[0]:.2f}")
print(f"Intercept: {reg.intercept_:.2f}")

# Manually calculate the price using the formula: Y = mX + b
manual_price_3300 = 3300 * reg.coef_[0] + reg.intercept_
print(f"Manually calculated price of a 3300 sq ft home: ${manual_price_3300:,.2f}")

manual_price_5000 = 5000 * reg.coef_[0] + reg.intercept_
print(f"Manually calculated price of a 5000 sq ft home: ${manual_price_5000:,.2f}")
```

```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('canada_per_capita_income.csv')

print(df.head())

X = df[['year']]
y = df['per_capita_income']

reg = LinearRegression()
reg.fit(X, y)


year_2020 = np.array([[2020]])
predicted_income_2020 = reg.predict(year_2020)
print(f"Predicted Per Capita Income in 2020: ${predicted_income_2020[0]:,.2f}")

plt.xlabel('Year')
plt.ylabel('Per Capita Income (in $)')
plt.scatter(df['year'], df['per_capita_income'], color='red', marker='+', label='Data Points')
plt.plot(df['year'], reg.predict(X), color='blue', label='Regression Line')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('salary.csv')

print("Initial Data:\n", df.head())

df.dropna(inplace=True)

print("\nData after handling missing values:\n", df.head())

X = df[['YearsExperience']]
y = df['Salary']

reg = LinearRegression()
reg.fit(X, y)

experience_12 = np.array([[12]])
predicted_salary_12 = reg.predict(experience_12)
print(f"\nPredicted Salary for 12 Years of Experience: ${predicted_salary_12[0]:,.2f}")


plt.xlabel('Years of Experience')
plt.ylabel('Salary (in $)')
plt.scatter(df['YearsExperience'], df['Salary'], color='green', marker='o', label='Data Points')
plt.plot(df['YearsExperience'], reg.predict(X), color='orange', label='Regression Line')
plt.legend()
plt.show()
```

```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('hiring.csv')

word_to_number = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9,
    'ten': 10, 'eleven': 11, 'twelve': 12
}

df['experience'] = df['experience'].replace(word_to_number)

df.fillna(0, inplace=True)

X = df[['experience', 'test_score)', 'interview_score']]
y = df['salary']

reg = LinearRegression()
reg.fit(X, y)

candidates = np.array([[2, 9, 6], [12, 10, 10]])
predicted_salaries = reg.predict(candidates)

print(f"Predicted Salary for 2 yrs exp, 9 test score, 6 interview score: ${predicted_salaries[0]:,.2f}")
print(f"Predicted Salary for 12 yrs exp, 10 test score, 10 interview score: ${predicted_salaries[1]:,.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('1000_Companies.csv')

label_encoder = LabelEncoder()
df['State'] = label_encoder.fit_transform(df['State'])

X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y = df['Profit']

reg = LinearRegression()
reg.fit(X, y)

state_code = label_encoder.transform(['Florida'])[0]
new_data = np.array([[91694.48, 515841.3, 11931.24, state_code]])
predicted_profit = reg.predict(new_data)

print(f"Predicted Profit for given data: ${predicted_profit[0]:,.2f}")
```
